1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def eq(t, x):
    return x/t + 1

t0 = 1
x0 = 0

t_span = [1, 10]

sol = solve_ivp(eq, t_span, [x0], t_eval=np.linspace(t_span[0], t_span[1], 100))

plt.plot(sol.t, sol.y[0])
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

2. Encuentre la solución numérica para un oscilador armónico simple:

$$
\frac{d^2 x}{dt^2} +  \omega_{0}^2 x = 0.
$$
Donde $x_0=0.5$, $\dot{x}_0=0.1$ y $\omega_0=3$.

2.1 Defina una función, cuyos argumentos de entrada sean:


1. el tiempo $t$,
2. el arreglo $[x,\dot{x}]$.

Y la salida:

1. el arreglo $[\dot{x},\ddot{x}]$.

  Considerando el argumento de la frecuencia natural $\omega_0$ debe ser una variable global.

```python
def harmonico(t, y, w):
    dydt = y[1]
    d2ydt2 = -w**2 * y[0]
    return [dydt, d2ydt2]

x0 = 0.5
x1 = 0.1
w = 3.0
y0 = [x0, x1]

t0 = 0.0
tf = 10.0
puntos = 1000

sol = solve_ivp(harmonico, [t0, tf], y0, t_eval=np.linspace(t0, tf, puntos), args=(w,))

plt.plot(sol.t, sol.y[0])
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.title('Oscilador Armónico Simple')
plt.show()
```
